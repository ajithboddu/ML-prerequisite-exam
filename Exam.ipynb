Q1

 Matrix multiplication between a (n, k) and (k, m) matrices involves nmk multiplications and nm(k-1) additions. To elaborate, each element in the resulting (n, m) matrix is obtained by performing k multiplications and k-1 additions between the corresponding elements of the first matrix row and the second matrix column. As there are nm elements in the result matrix, the total number of multiplications is nmk, and the total number of additions is nm(k-1).

```python
import numpy as np
import time

# Define matrix dimensions
rows = int(200)
cols_A = int(500)
cols_B = int(600)

# Generate random matrices A and B using NumPy
matrix_A = np.random.rand(rows, cols_A) 
matrix_B = np.random.rand(cols_A, cols_B)

# Initialize result matrices for both approaches
result_matrix_list_of_lists = [[0 for _ in range(cols_B)] for _ in range(rows)]
result_matrix_numpy = np.zeros((rows, cols_B))

# Matrix multiplication using a list of lists (traditional approach)
start_time = time.time()
for i in range(rows):
    for j in range(cols_B):
        for x in range(cols_A):
            result_matrix_list_of_lists[i][j] += matrix_A[i][x] * matrix_B[x][j]
end_time = time.time()
time_taken_list_of_lists = end_time - start_time
print("Time taken (List of Lists): ", time_taken_list_of_lists)         

# Matrix multiplication using NumPy's dot function
start_time = time.time()
result_matrix_numpy = np.dot(matrix_A, matrix_B)
end_time = time.time()
time_taken_numpy = end_time - start_time
print("Time taken (NumPy): ", time_taken_numpy)   

# Comparison of timings
if time_taken_list_of_lists < time_taken_numpy:
    print("The traditional approach using a list of lists is faster.")
elif time_taken_list_of_lists > time_taken_numpy:
    print("NumPy's dot function is faster.")
else:
    print("Both methods took the same amount of time.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting functions with various mathematical relationships

# Function 1: y = x
x1 = np.arange(0.5, 100, 0.5)
y1 = x1
fig1, ax1 = plt.subplots()
ax1.plot(x1, y1)
ax1.set(xlabel='x', ylabel='y', title="1. Linear Relationship: y = x")
ax1.grid()
plt.show()

# Function 2: y = x^2
x2 = np.arange(0.5, 100, 0.5)
y2 = x2**2
fig2, ax2 = plt.subplots()
ax2.plot(x2, y2)
ax2.set(xlabel='x', ylabel='y', title="2. Quadratic Relationship: y = x^2")
ax2.grid()
plt.show()

# Function 3: y = x^3/100
x3 = np.arange(0.5, 100, 0.5)
y3 = (x3**3) / 100
fig3, ax3 = plt.subplots()
ax3.plot(x3, y3)
ax3.set(xlabel='x', ylabel='y', title="3. Cubic Relationship: y = x^3/100")
ax3.grid()
plt.show()

# Function 4: y = sin(x)
x4 = np.arange(0.5, 100, 0.5)
y4 = np.sin(x4)
fig4, ax4 = plt.subplots()
ax4.plot(x4, y4)
ax4.set(xlabel='x', ylabel='y', title="4. Sine Function: y = sin(x)")
ax4.grid()
plt.show()

# Function 5: y = sin(x)/x
x5 = np.arange(0.5, 100, 0.5)
y5 = np.sin(x5) / x5
fig5, ax5 = plt.subplots()
ax5.plot(x5, y5)
ax5.set(xlabel='x', ylabel='y', title="5. Sinusoidal Decay: y = sin(x)/x")
ax5.grid()
plt.show()

# Function 6: y = log(x)
x6 = np.arange(0.5, 100, 0.5)
y6 = np.log(x6)
fig6, ax6 = plt.subplots()
ax6.plot(x6, y6)
ax6.set(xlabel='x', ylabel='y', title="6. Logarithmic Relationship: y = log(x)")
ax6.grid()
plt.show()

# Function 7: y = e^x
x7 = np.arange(0.5, 100, 0.5)
y7 = np.exp(x7)
fig7, ax7 = plt.subplots()
ax7.plot(x7, y7)
ax7.set(xlabel='x', ylabel='y', title="7. Exponential Growth: y = e^x")
ax7.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Generate a 20x5 matrix of random numbers between 1 and 2
random_matrix = np.random.uniform(1, 2, size=(20, 5))

# Create a DataFrame using Pandas
df = pd.DataFrame(random_matrix, columns=["a", "b", "c", "d", "e"])

# Add a new column "f" which is the sum of columns "a", "b", "c", "d", "e"
df["f"] = df["a"] + df["b"] + df["c"] + df["d"] + df["e"]

# Create a new column "g" based on conditions
df["g"] = np.where(df["f"] < 8, "LT8", "GT8")

# Find the number of rows where the value in column "g" is "LT8"
lt8_rows_count = df[df["g"] == "LT8"].shape[0]

# Find the standard deviation of column "f" for rows where the value in column "g" is "LT8" and "GT8" respectively
std_lt8 = df[df["g"] == "LT8"]["f"].std()
std_gt8 = df[df["g"] == "GT8"]["f"].std()

# Print the results
print("Number of rows where value in column 'g' is 'LT8':", lt8_rows_count)
print("Standard deviation of column 'f' for 'LT8':", std_lt8)
print("Standard deviation of column 'f' for 'GT8':", std_gt8)
```

```python
import numpy as np

def custom_argmin(arr):
    min_index = 0  # Assume the minimum index is initially 0
    min_value = arr[0]  # Assume the minimum value is initially the first element

    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i

    return min_index

# Test the custom_argmin function
array_to_test = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])
result_custom = custom_argmin(array_to_test)
result_np = np.argmin(array_to_test)

print("Custom argmin result:", result_custom)
print("NumPy argmin result:", result_np)
```

```python
import jax
import jax.numpy as jnp

# Define a JAX function and its analytical counterpart
def jax_function(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

def analytical_function(x, y):
    real_part = 2 * x * y + (y**3) * jnp.cos(x)
    imag_part = x**2 + 3 * (y**2) * jnp.sin(x)
    result = complex(real_part, imag_part)
    return result

# Compute the gradient using JAX and evaluate the analytical solution
gradient_function = jax.grad(jax_function, argnums=(0, 1))
x_value = np.random.rand()
y_value = np.random.rand()
gradient_at_point = gradient_function(x_value, y_value)
analytical_solution = analytical_function(x_value, y_value)

# Print the results
print("Gradient at a random point using JAX:", complex(gradient_at_point[0], gradient_at_point[1]))
print("Analytical Solution:", analytical_solution)
```
